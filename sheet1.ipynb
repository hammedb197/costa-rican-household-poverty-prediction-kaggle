```python
train = train.drop(['dependency', 'edjefe', 'edjefa', 'dependency', 'v2a1', 'idhogar', 'rez_esc'], axis = 1)
```

```python
X = train.iloc[:, 1:-1]
y = train.iloc[:, -1:]
```

```python
# y
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
# y = scale.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X ,y, test_size =0.25, random_state= 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

algo = RandomForestClassifier()
params={'n_estimators': list(range(40,61, 1))}
algo_cv = GridSearchCV(algo, params, cv=5)
algo_cv.fit(x_train, y_train)
y_pred = algo_cv.predict(x_test)
```

```python
from sklearn.metrics import  accuracy_score, confusion_matrix, classification_report

print("Classification report ", classification_report(y_pred, y_test))
```

```python
print("Confusion matrix", confusion_matrix(y_pred, y_test))
```

```python
print("accuracy", accuracy_score(y_pred, y_test))
```

```python
test = test.fillna(0)
```

```python
Id = test.index
```

```python
def preprocess(data):
    data = data.fillna(0)
    data = data.drop(['dependency', 'edjefe', 'edjefa', 'dependency', 'v2a1', 'idhogar', 'rez_esc'], axis = 1)
    data = data.iloc[:, 1:]
    scale = StandardScaler()
    data = scale.fit_transform(data)
    
    
    return data
  
```

```python
test = preprocess(test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

algo = RandomForestClassifier()
params={'n_estimators': list(range(40,61, 1))}
algo_cv = GridSearchCV(algo, params, cv=5)
algo_cv.fit(X, y)
y_pred = algo_cv.predict(test)
```

```python
algo_cv.score(test, y_pred)
```

```python
pred = pd.DataFrame({'Id' : Id, "Target" : y_pred})
```
